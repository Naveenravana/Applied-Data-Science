```python
def generateX(N):
    x = np.random.random(N)*2*np.pi
    yd = np.sin(x)
    return x, yd
```

```python
x, y = generateX(500)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
def plotmodel(x, y, yd):
    i = x.argsort()
    plt.figure()
    plt.plot(x[i], y[i], "g-o")
    plt.plot(x[i], yd[i],  "r-o")
    plt.ylabel("F(X)")
    plt.xlabel("X")
    plt.legend(["estimated", "True"])
    plt.title("Comparision")
    plt.show()
```
